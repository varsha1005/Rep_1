**Create a pandas dataframe(Dataframe name as 'df')with numpy random values(4 features and 4 observations)**

```python
import pandas as pd
import numpy as np
import random
alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
data={
    'Cosmetics':[random.choice(list(alphabet)) for _ in range(4)],
    'Color':[random.choice(list(alphabet)) for _ in range(4)],
    'Cost':np.random.randint(100,8000,size=4),
    'Usage %':np.random.uniform(0.0,100.0,4),

}
df=pd.DataFrame(data)
df
```

**Rename the task-1 'df' dataframe column names to 'Random value 1','Random value 2','Random value 3','Random value 4'**

```python
df.rename(columns={'Cosmetics':'Random value 1','Color':'Random value 2','Cost':'Random value 3','Usage %':'Random value 4'},inplace=True)
df
```
